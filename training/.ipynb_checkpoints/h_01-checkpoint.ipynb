```python
from tkinter import *
from tkinter import messagebox
 
root = Tk()
 
# 버튼 클릭 이벤트 핸들러
def okClick():
    name = txt.get()
    messagebox.showinfo("이름", name)
 
lbl = Label(root, text="이름")
lbl.grid(row=0, column=0)
txt = Entry(root)
txt.grid(row=0, column=1)
 
# 버튼 클릭 이벤트와 핸들러 정의
btn = Button(root, text="OK", command=okClick)
 
btn.grid(row=1, column=1)
 
root.mainloop()
```

```python
from tkinter import *
from tkinter import ttk

# Tkinter 인스턴스 생성
win = Tk()

# Window 사이즈 지정
win.geometry("300x200+100+100")
win.resizable(False, False)

# get() : entry에 텍스트를 문자열로 반환
def Cal():
    a = int(entry.get())
    if a %2 == 1 :
        label.config(text='홀수')
    else :
        label.config(text='짝수')


entry = Entry(win)
entry.pack()

button = Button(win, text="계산", command=Cal)
button.pack()

label = Label(win)
label.pack()

win.mainloop()
```

```python
from tkinter import *
import random

root = Tk()
root.geometry('300x200')
root.config(bg='#eee')

def ClickLotto():
    lotto = random.sample(range(1,46),6) 
#     lotto.sort(reverse=True)
    lotto.sort()
    lbl.config(text=lotto) 
    
lbl = Label(root)
lbl.config(relief='ridge',font=("",8,""))
lbl.place(x=10,y=10,width=200,height=20)

btn = Button(root)
btn.config(text='Lotto',command=ClickLotto)
btn.place(x=10,y=30,width=200,height=20)

root.mainloop()

```

```python
from tkinter import*
from tkinter import ttk
from functools import partial


root = Tk()
root.geometry('300x200')
root.config(bg='#eee')
global m_num
m_num=0


def click(key):
    global m_num
    if key=='+':
        m_num+=1
    elif key =='-':
        m_num-=1
    else:
        m_num=0
    lbl.config(text=m_num)
        
# def ClickDown(key):
#     global m_num
#     if key=='-':
#         m_num-=1
#     else:
#         m_num=0

        
lbl = Label(root)
lbl.config(relief='ridge',font=("",8,""))
lbl.place(x=10,y=10,width=100,height=20)

btn1 = Button(root)
btn1.config(text='+',command=partial(click,'+'))
btn1.place(x=10,y=30,width=100,height=40)

btn2 = Button(root)
btn2.config(text='-',command=partial(click,'-'))
btn2.place(x=10,y=70,width=100,height=40)

root.mainloop()
```

```python
from tkinter import *
from tkinter import ttk

incr=Tk()
incr.title("increasing_numbers")

global m_num
m_num=0
#버튼 명령
def click(key):
    global m_num
    if key=='+':
        m_num+=1
    elif key=='-':
        m_num-=1
    else:
        m_num=0
    num_var.set(m_num)

#라벨 생성
num_var=IntVar()
w=Label(incr, textvariable=num_var)
w.grid(row=0)

#+,-,리셋 버튼생성
operator_list=['+','-','reset']

r=1
for btn_text in operator_list:
    def cmd(x=btn_text):
        click(x)
    ttk.Button(incr,text=btn_text,command=cmd).grid(row=r)
    r+=1


incr.mainloop()
```
