```python
from tkinter import*
from tkinter import ttk
from functools import partial


root = Tk()
root.geometry('300x160')
root.config(bg='#eee')
m_num=0

def click(key):
    global m_num
    if key=='+':
        m_num+=1
    elif key =='-':
        m_num-=1
    else:
        m_num=0
    lbl.config(text=m_num)
        
        
lbl = Label(root)
lbl.config(relief='ridge',font=("",8,""))
lbl.place(x=100,y=10,width=100,height=20)

btn1 = Button(root)
btn1.config(text='+',command=partial(click,'+'))
btn1.place(x=100,y=30,width=100,height=40)

btn2 = Button(root)
btn2.config(text='-',command=partial(click,'-'))
btn2.place(x=100,y=70,width=100,height=40)

btn3 = Button(root)
btn3.config(text='Reset',command=partial(click,'Reset'))
btn3.place(x=100,y=110,width=100,height=40)

root.mainloop()
```

```python
from tkinter import*
from tkinter import ttk

win = Tk()
win.geometry('500x500')


cnt=0

def numAdd():
    global cnt
    cnt = cnt + 1
    lbl.config(text=cnt)


def numSub():
    global cnt
    cnt = cnt - 1
    lbl.config(text=cnt)
    
    
lbl = Label(win)
lbl.config(text='?',justify=CENTER, relief=RIDGE)
lbl.place(x=10, y=10,width=100,height=50)

btnp = Button(win)
btnp.config(text="+",font=("",8,""),command=numAdd)
btnp.place(x=10, y=70, width=100,height=50)
            
btnp2 = Button(win)
btnp2.config(text="-",font=("",8,""),command=numSub)
btnp2.place(x=10, y=130, width=100,height=50)

win.mainloop()
```

```python
from tkinter import *
from tkinter import ttk

incr=Tk()
incr.title("increasing_numbers")

m_num=0
#버튼 명령
def click(key):
    global m_num
    if key=='+':
        m_num+=1
    elif key=='-':
        m_num-=1
    else:
        m_num=0
    num_var.set(m_num)

#라벨 생성
num_var=IntVar()
w=Label(incr, textvariable=num_var)
w.grid(row=0)

#+,-,리셋 버튼생성
operator_list=['+','-','reset']

r=1
for btn_text in operator_list:
    def cmd(x=btn_text):
        click(x)
    ttk.Button(incr,text=btn_text,command=cmd).grid(row=r)
    r+=1


incr.mainloop()
```

```python
from tkinter import *
from tkinter import ttk
import random
# 윈도우 창 설정
win = Tk()
win.title('숫자찾기게임')
win.geometry('330x150')

# 변수설정
num = 0
betNum = random.randint(1,100)
cnt = 0

# 함수 설정
def NumGame():
    global num
    global betNum
    global cnt
    cnt+=1
    btnp['text'] = f'클릭 했습니다.'
    lblcnt.config(text=f'횟수 : {cnt}')
    
    num = int(entry.get())
    if num < betNum:
        lbl2['text'] = num
    elif num > betNum:
        lbl3['text'] = num
    else:
#         lbl1.config(text=f'정답: {num}')
        lbl1['text'] = f'정답 : {num}'
    
        
def reset():
    global num
    global betNum
    global cnt
    
    num = 0
    cnt = 0
    
    entry.delete(0,END)
    lbl1.config(text="")
    lbl2.config(text="0")
    lbl3.config(text="100")
    lblcnt.config(text="횟수")
    betNum = random.randint(1,100)
    btnp['text'] = "클릭"

            
        
# 버튼생성    

lblname1 = Label(win)
lblname1.config(text='COM',justify=CENTER)
lblname1.place(x=10, y=10,width=100,height=25)

lblname2 = Label(win)
lblname2.config(text='작은수',justify=CENTER)
lblname2.place(x=115, y=10,width=100,height=25)

lblname3 = Label(win)
lblname3.config(text='큰수',justify=CENTER)
lblname3.place(x=220, y=10,width=100,height=25)
    
lbl1 = Label(win)
lbl1.config(text='?',justify=CENTER, relief=RIDGE)
lbl1.place(x=10, y=35,width=100,height=25)

lbl2 = Label(win)
lbl2.config(text='0',justify=CENTER, relief=RIDGE)
lbl2.place(x=115, y=35,width=100,height=25)

lbl3 = Label(win)
lbl3.config(text='100',justify=CENTER, relief=RIDGE)
lbl3.place(x=220, y=35,width=100,height=25)

lbl4 = Label(win)
lbl4.config(text='입력수치',justify=CENTER)
lbl4.place(x=10, y=75,width=100,height=25)

entry = Entry(win)
entry.config(text="",font=("",8,""))
entry.place(x=115, y=75,width=100,height=25)

btnp = Button(win)
btnp.config(text="확인",font=("",8,""),command=NumGame)
btnp.place(x=220, y=75,width=100,height=25)

rebtn = Button(win)
rebtn.config(text="게임을 다시시작",font=("",8,""),command=reset)
rebtn.place(x=220, y=110,width=100,height=25)

lblcnt = Label(win)
lblcnt.config(text='횟수',justify=CENTER)
lblcnt.place(x=10, y=110,width=100,height=25)

win.mainloop()
```
