```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
# import image
im = cv.imread('son_pic.jpg')
```

```python
# BGR
plt.figure()
plt.imshow(im)
plt.title('Original')
```

```python
#RGB
rgb = cv.cvtColor(im, cv.COLOR_BGR2RGB)
plt.figure()
plt.imshow(rgb)
plt.title('RGB')
```

```python
# gray scale
gray = cv.cvtColor(im,cv.COLOR_BGR2GRAY)
plt.figure()
plt.imshow(gray,cmap='gray')
plt.title('Gray')
```

```python
# blur
blur = cv.blur(im,(15,15))
blur = cv.cvtColor(blur,cv.COLOR_BGR2RGB)
plt.subplot(121)
plt.imshow(rgb)
plt.title('RGB')
plt.subplot(122)
plt.imshow(blur)
plt.title('Blur')
```

```python
# edge detection
edges=cv.Canny(gray,0,100)
plt.subplot(121)
plt.imshow(gray,cmap='gray')
plt.title('gray')
plt.subplot(122)
plt.imshow(edges,cmap='gray')
plt.title('Edge Detection')
```
