# Unemployment Rate

# filter 'Unemployment Rate

```python
# filter 'Unemployment Rate
df_ur = df[df['Subject Descriptor'].str.contains('Unemployment')][['Country','2021']]
df_ur
```

```python
#check info
df_ur.info()
```

```python
#filter out Nan rows
df_ur = df_ur.loc[df_ur['2021'].notnull()]
df_ur
```

```python
# reset row index
df_ur.reset_index(drop=True,inplace=True)
df_ur
```

```python
# check info
df_ur.info()
```

```python
# convert to numeric data
df_ur['2021']=pd.to_numeric(df_ur['2021'])
df_ur
```

```python
# check info
df_ur.info()
```

```python
#plot bar graph
df_ur.plot.bar(x='Country')
```

```python
#sort and plot
df_ur.sort_values('2021',ascending=False).plot.bar(x='Country',title='Unemploytment Rate',figsize=(15,5))
df_ur
```

```python
# to numpy
df_ur_numpy =df_ur.sort_values('2021',ascending=False).to_numpy()
```

```python
print(df_ur_numpy)
```

```python
# import numpy and matploblib
import matplotlib.pyplot as plt
import numpy as np
```

```python
# plot
plt.rcParams['figure.figsize']=(20,3)
plt.xticks(rotation='vertical')
plt.bar(df_ur_numpy[:,0],df_ur_numpy[:,1])
plt.title('2021 Unemployment Rate - IMF World EconomiOutlook Database,April 2020')
# plog Korea data again
idx=np.where(df_ur_numpy=='Korea')
plt.bar(df_ur_numpy[idx[0],0],df_ur_numpy[idx[0],1],label='Korea')
# plog USA data again
idx=np.where(df_ur_numpy=='United States')
plt.bar(df_ur_numpy[idx[0],0],df_ur_numpy[idx[0],1],label='United States')
# plog China data again
idx=np.where(df_ur_numpy=='China')
plt.bar(df_ur_numpy[idx[0],0],df_ur_numpy[idx[0],1],label='Chinaea')
# plog Japen data again
idx=np.where(df_ur_numpy=='France')
plt.bar(df_ur_numpy[idx[0],0],df_ur_numpy[idx[0],1],label='France')
plt.legend()
```

# Add new column and group analysis

```python
df_ur
```

```python
del(df_ur['criteria'])
df_ur
```

```python
# update criteria to low, medium, and high
df_ur.loc[df_ur['2021'] < 5,'Criteria'] = 'Low'
df_ur.loc[df_ur['2021'] >= 5 & (df_ur['2021'] <10),'Criteria'] = 'Medium'
df_ur.loc[df_ur['2021'] >=10, 'Criteria'] = 'High'
df_ur
```

```python
#find mean for each criteria
df_ur.groupby(['Criteria']).mean()
```
