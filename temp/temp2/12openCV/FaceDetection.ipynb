```python
#import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load image
img = cv2.imread('./spurs_pic.jpg')
plt.imshow(img)
```

```python
# convert RGB
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb)
```

```python
# convert tp grausca;e
gray=cv2.cvtColor(rgb, cv2.COLOR_RGB2GRAY)
plt.imshow(gray,cmap='gray')
```

```python
# face detection
classfier=cv2.CascadeClassifier(('.\\haarcascades\\haarcascade_frontalface_default.xml'))
```

```python
rects=classfier.detectMultiScale(
    gray,
    scaleFactor=1.2,
    minNeighbors=5
)
```

```python
print('Faces found: {}'.format(len(rects)))
```

```python
#rectangle
# x,y,w,h = rects[0]
for(x,y,w,h) in rects:
    # draw a rectangle
    cv2.rectangle(rgb, (x,y), (x+w,y+h), (0,255,0),2)
```

```python
cv2.rectangle(rgb, (x,y), (x+w,y+h), (0,255,0),2)
plt.imshow(rgb)
```

```python
# save pic
bgr = cv2.cvtColor(rgb,cv2.COLOR_RGB2BGR)
cv2.imwrite('spurs_pic_faces.jpg',bgr)
```
