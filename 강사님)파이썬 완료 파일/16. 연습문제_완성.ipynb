### 1~100까지 자연수에서 3의배수, 5의 배수를 각각 구하고, 3또는 5의 배수에 해당하는 수들의 총 합계를 구하시오.
- 주의: 3과 5의 공배수(예:15)를 중복으로 더하지 않도록 조심한다.


```python
num3 = []
num5 = []
result=0

for i in range(1,101):
    if (i%3 ==0) or (i%5 == 0) :
        result += i
        if i % 3 == 0 : num3.append(i)
        if i % 5 == 0 : num5.append(i)
print(f'1~100 사이에서 3의 배수 : {str(num3)[1:-1]}')
print(f'1~100 사이에서 5의 배수 : {str(num5)[1:-1]}')
print(f'1~100 사이에서 3 또는 5의 배수의 합계: {result}')
    
```

### 아래 조건을 만족하는 코드를 완성하시오.

<출력화면>


```python
def mul100(n):
    return n*100
```

```python
num_list = list(range(51))

for i in num_list:
    if i % 3 == 0 : num_list[i]= mul100(i)
        
print(num_list)
```

### 아래의 덧셈게임을 프로그래밍 하시오
- 컴퓨터가 무작위로 생성한 세자리 덧셈문제를 사용자가 암산으로 맞히는 프로그램
- 시작 : 아무키나, 종료 : 0 을 사용자 입력으로 받고 그에 따라 프로그램 실행 
- 정답이면 +10점, 오답이면 -5점
- 마지막에 '당신의 점수는 (점수)점입니다' 출력


```python
import random

score = 0
print ('지금부터 덧셈게임을 시작 하겠습니다.')

while True :
    data = input('시작 : 아무키나, 종료 : 0 --> ')
    if data == '0' : break
    
    num1 = random.randint(1,999)
    num2 = random.randint(1,999)
    ans = int(input(f'{num1}+{num2}= '))
    
    if ans == (num1+num2) :
        score += 10
        print(f'정답!!!!, score : {score}점')
        
    else :
        score -= 5
        print(f'오답!!!!, score : {score}점')
         
print(f'당신의 덧셈점수는 {score}점 입니다.')

```

### 아래 질문에 알맞은 프로그램을 작성하시오.
- 친환경 열풍을 타고 2022년에 ICT자동차에서 전기차 supersonic 을 출시하였습니다.
- ICT자동차에서는 출시 이벤트로 선착순 50대에 한하여 supersonic 로고가 박힌 골프백를 증정하고자 합니다.
- 이에 전국의 각 영업점에서 증정품 재고수량을 파악할 수 있는 system을 구축하고자 합니다.
- 시스템 코드를 확인하고 적절한 예외처리 구문을 넣으시오.

- 조건1 : 1보다 작거나 숫자가 아닌 입력값이 들어올 때는 ValueError 로 처리(출력 메세지 : "잘못된 값을 입력하였습니다.")
- 조건2 : 골프백 소진시 사용자 정의 에러[SoldOutError]를 발생시키고 프로그램 종료(출력 메세지 : '선착순 선물 재고가 소진되어 이벤트를 종료합니다.')
- 골프백 수량만큼 ♥(\u2665) 출력(예시 참조)


```python
class SoldOutError(Exception): 
    pass

gift = 50

while True:
    try:
        print(f"[남은 골프백 수량 : {gift}]")
        order = int(input("supersonic 판매수량을 입력하세요."))
        if order > gift :
        
            print("골프백이 부족합니다.")
        elif order < 1:
            raise ValueError    
        else:
            print(f"supersonic {order}대, 골프백 ", end='')
            print('\u2665'*order,'주문이 완료되었습니다.')
            gift -= order

        if gift == 0: 
            raise SoldOutError
    
    except ValueError:
        print("잘못된 값을 입력하였습니다.")
    
    except SoldOutError: 
        print('선착순 선물 재고가 소진되어 이벤트를 종료합니다.')
        break
        
```

### 아래 질문에 알맞은 프로그램을 작성하시오.
- 당신은 음식점의 계산을 도와주는 점원이다. 
- 카운터에는 거스름돈으로 사용할 500원, 100원, 50원, 10원 짜리의 동전이 무한히 존재한다고 가정한다.

#### 손님에게 거슬러 줘야 할 돈이 1260원일때 거슬러 줘야 할 동전의 최소 개수를 구하라.
- 출력 -> 거스름돈 : 1260원, 거슬러 줘야 할 동전의 개수 : 6개


```python
change = 1260
count = 0
coins = [500, 100, 50, 10]

print(f'거스름돈 : {change}원, ', end='')
for coin in coins :
    count += change // coin
    change %= coin
    
print(f'거슬러 줘야 할 동전의 개수 : {count}개')
```
